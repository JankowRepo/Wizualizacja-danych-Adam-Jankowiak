```python
import plotly.express as px
import pandas as pd
import numpy as np
from plotly.offline import init_notebook_mode
init_notebook_mode(connected=True)
```

```python
df = pd.read_csv('C:/Users/Adam/Downloads/zamowienia.csv', sep=';')
df.head()
```

```python
# Zadanie 1
utarg=df.where(df["Kraj"]=="Polska").groupby("Sprzedawca").sum('Utarg')["Utarg"]
fig = px.funnel(utarg)
fig.show()
```

```python
# Zadanie 2
fig = px.density_contour(df, x="Utarg", y="Sprzedawca")
fig.update_traces(contours_coloring="fill")
fig.show()
```

```python
# Zadanie 3
df2 = pd.read_csv('C:/Users/Adam/Downloads/FIFA22_official_data.csv')
df2.head()
```

```python
df2_data = df2[df2['Nationality'].isin(['Poland', 'Spain'])]
df2_data["Value"].head()
```

```python
df2_data["Value"] = df2_data["Value"].str.replace('€', '')
df2_data["Value"] = df2_data["Value"].fillna(0)
df2_data["Value"] = df2_data['Value'].replace({'K': '*1e3', 'M': '*1e6'}, regex=True).map(pd.eval).astype(float)
df2_data.dropna(subset=['Club'])
df2_data["Value"]
```

```python
fig = px.icicle(df2_data, path=['Name', 'Nationality', 'Club'], values='Value')
fig.update_traces(root_color="darkgreen")
fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()
```
