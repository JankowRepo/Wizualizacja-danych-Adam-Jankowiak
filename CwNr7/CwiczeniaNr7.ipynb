```python
# Zadanie 1
df = pd.read_csv('C:/Users/Adam/Desktop/repo/wd_dswp/lab_7/data/FIFA22_official_data.csv')
df
```

```python
df=df[['Name', "GKDiving", "GKHandling", "GKKicking", "GKPositioning", "GKReflexes"]]
df=df[22:24]
df
```

```python
def fun(df):
    df=df.iloc[:,:100]
    etykiety=df.iloc[:,1:].columns
    angles = np.linspace(0, 2 * np.pi, len(etykiety), endpoint=False).tolist()
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
    
    for i in range(len(df)):
        rgb = np.random.rand(3,)
        ax.fill(angles, df.iloc[i,1:].values.flatten(), c=rgb, alpha=0.25, label=df.iloc[i,:]["Name"])
        
    ax.set_thetagrids(np.degrees(angles), etykiety)
    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)
    plt.legend()
    plt.show
        
fun(df)
```

```python
import matplotlib.animation as animation

df = pd.read_csv('C:/Users/Adam/Desktop/repo/wd_dswp/lab_7/data/pkn.txt')
df.head(20)

fig, ax = plt.subplots()

x = np.arange(0, 20, 1)
y = df['<CLOSE>'][:20]
line, = ax.plot(x, y)


def animate(i):
    X = np.arange(i, i + 20, 1)
    Y = df['<CLOSE>'][i:20 + i]
    line.set_data(X, Y)
    ax.set_xlim(i, i + 25)
    return line,


ani = animation.FuncAnimation(
    fig, animate, interval=500, blit=False, save_count=50)
plt.ylim(12, 21)
plt.show()
```
