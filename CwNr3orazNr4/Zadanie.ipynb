```python
p0 = df.BMI.min()
p100 = df.BMI.max()
q1 = df.BMI.quantile(0.25)
q2 = df.BMI.quantile(0.5)
q3 = df.BMI.quantile(0.75)
iqr = q3-q1
p0, p100, q1, q2, q3, iqr
```

```python
lc = q1 - 1.5* iqr
uc = q3 + 1.5 * iqr
lc, uc
```

```python
df.BMI.plot(kind='box')
```

```python
df.BMI.clip(upper=uc, inplace=True)
df.BMI.clip(lower=lc, inplace=True)
df.BMI.plot(kind='box')
```
