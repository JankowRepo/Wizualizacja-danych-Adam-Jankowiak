```python
# Praca zawiera opracowania na temat zabytków nieruchomych w Polsce
# Link do datasetu: https://dane.gov.pl/pl/dataset/1130,rejestr-zabytkow-nieruchomych
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ZESTAWIENIE_ZABYTKÓW_NIERUCHOMYCH.csv', sep=';')
df.head(15)
```

```python
df.describe()
```

```python
data=df[['INSPIRE_ID', 'WOJEWODZTWO']].groupby('WOJEWODZTWO').count()['INSPIRE_ID'].sort_values(ascending=False)

plt.figure(figsize=(10, 10))
plt.pie(data,pctdistance=1.1,  autopct=lambda pct: "{:.1f}%".format(pct), textprops=dict(color="black"), startangle=90)
plt.title("Procentowa ilość zarejestrowanych zabytków w Polsce\n z podziałem na województwa")
plt.legend(data.index,
          title ="Województwa",
          loc ="center left",
          bbox_to_anchor =(1.1, 0, 0.5, 1))
plt.show
```

```python
data=df[['INSPIRE_ID', 'POWIAT']].groupby('POWIAT').count()['INSPIRE_ID'].sort_values(ascending=False)
my_list=[(str(round(data.values[i]/df["INSPIRE_ID"].count()*100,1))+"%") for i in range(10)]

plt.figure(figsize=(10, 10))
plt.pie(data[:10],labels=my_list, textprops=dict(color="black"), rotatelabels=45, startangle=90)
plt.title("Procentowa ilość zarejestrowanych zabytków w 10 najpopularniejszych pod tym względem miast w Polsce")
plt.legend(data.index[:10],
          title ="Województwa",
          loc ="center left",
          bbox_to_anchor =(1.1, 0, 0.5, 1))
plt.show
```

```python
plt.figure(figsize=(12, 8))
ax = sns.histplot(df.DOKLADNOSC_POLOZENIA)
ax.bar_label(ax.containers[0])

plt.title("Podsumowanie na temat wiedzy o dokładności położenia zabytów w Polsce")
plt.xlabel("Dokładność położenia zabytku")
plt.ylabel("Ilość zabytków")
plt.show()
```

```python
df["Years"]=pd.DatetimeIndex(pd.to_datetime(df['DATA_WPISU'])).year
df["Months"]=pd.DatetimeIndex(pd.to_datetime(df['DATA_WPISU'])).month
data=df[['INSPIRE_ID', 'Months']].groupby('Months').count()['INSPIRE_ID']

plt.figure(figsize=(12, 8))
plt.stem(data)
plt.title("Ilość rejestracji zabytków w każdym miesiącu przez cały czas trwania zestawienia")
plt.xlabel("Number miesiąca")
plt.ylabel("Łączna ilość rejestracy w danym miesiącu")
plt.show
```

```python
plt.figure(figsize=(14, 7))
plt.scatter(x = df.Years, y = df.index, s = 2)
plt.title("Wykaz rejestracji każdego zabytku wykonanej w ciągu ostatnych 122 lat")
plt.xlabel("Rok rejestracji")
plt.ylabel("Number zabytku")
plt.show
```

```python
plt.figure(figsize=(12, 7))

data=df[['INSPIRE_ID', 'NAZWA']].groupby('NAZWA').count()['INSPIRE_ID'].sort_values(ascending=False)[:10]
ax = sns.barplot(data.index , data.values)
ax.bar_label(ax.containers[0])

plt.xticks(rotation=30)
plt.title("Wykaz 10 najpopularniejszych rodzajów rejestrowanych zabytków")
plt.ylabel("Ilość rejestracji")
plt.xlabel("Nazwa zabytku")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
data= df.where(df.Years>df.Years.max()-20)[['INSPIRE_ID', 'Years']].groupby('Years').count()['INSPIRE_ID']

plt.fill_between(
        x = list(data.index.astype(int)),
        y1 =  data.values,
        where = data.values >= 700,
        color= "g",
        alpha= 0.9)
plt.fill_between(
        x = list(data.index.astype(int)),
        y1 =  data.values,
        where = data.values <= 700,
        color= "r",
        alpha= 0.9)


plt.title("Zestawienie rocznych ilości rejestracji w ciągu ostatnich 20 lat\n Kolorem ziolnym oznaczono lata z ponadprzeciętną liczbą rejestracji")
plt.ylabel("Ilość rejestracji")
plt.xlabel("Rok")
plt.show()
```

```python
plt.figure(figsize=(12, 8))

data=df[['INSPIRE_ID', 'FUNKCJA']].groupby('FUNKCJA').count()['INSPIRE_ID'].sort_values(ascending=False)[:10]
ax = sns.barplot(data.index , data.values)
ax.bar_label(ax.containers[0])

plt.xticks(rotation=40)
plt.title("Wykaz 10 najpopularniejszych funkcji rejestrowanych zabytków")
plt.ylabel("Ilość rejestracji")
plt.xlabel("Funkcja zabytku")
plt.show()
```
