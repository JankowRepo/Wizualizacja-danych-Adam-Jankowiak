```r
library("ggplot2")
library("readxl")
df <-read_excel('C:/Users/Adam/Desktop/Magisterka info/Data1/MyopiaDS1.xls')
head(df)
```

```r
# Sprawdzenie wartości odstających dla kolumny "waga" u dzieci w wieku szkolnym
boxplot(df['weight'])
```

```r
# Łączna średnia ilość czasu (w minutach) spędzonego na zewnątrz w ciągu tygodnia u dzieci w wieku szkolnym
hist(df$"time outdoors at weekend days" + df$"time outdoors at weekdays")
```

```r
# Zależność pomiędzy wzrostem a wadą wzroku u dzieci w wieku szkolnym (można dostrzec 
# niewielki wzrost wystąpień wyższej krótkowzroczności u dzieci mających powyżej 150cm wzrostu)
plot(df$height, df$"cylindrical refraction of right eye")
```

```r
# Korelacja między wzrostem a wagą u dzieci w wieku szkolnym. Na wykresie widać silną korelację co jest dosyć oczywiste
ggplot(df, aes(x=height, y=weight)) +
  geom_point() +
  geom_smooth(method=lm , color="red", fill="#69b3a2", se=TRUE) 
```

```r
# Zagęszczenie występowania wady wzroku u dzieci. Krótkowzroczność najczęściej wynosi od -0.75 do -0.25 dioptrii
d <- density(df$"cylindrical refraction of right eye")
plot(d, main="Density of cylindrical refraction")
polygon(d, col="red", border="blue") 
```
