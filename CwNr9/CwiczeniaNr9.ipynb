```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Zadanie 1
values = np.random.randn(100).cumsum()
plt.axvline(np.argmax(values), color='r')
sns.lineplot(data=values);
plt.axhline(y=0, xmin=0.0, xmax=1.0, color='w')
plt.axhline(y=np.mean(values), xmin=0.0, xmax=1.0, color='b', linestyle="--")
```

```python
# Zadanie 2 
values = np.random.randn(100).cumsum()
fig = sns.lineplot(data=values)
ax = fig.axes
ax.annotate("min",
            xy=(np.argmin(values), np.min(values)), xycoords='data',
            xytext=(len(values)/4, np.mean(values)), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            );
```

```python
# Zadanie 3
values = np.random.randn(100).cumsum()
fig = sns.lineplot(data=values)
ax = fig.axes
ax.annotate("min",
            xy=(np.argmin(values), np.min(values)), xycoords='data',
            xytext=(len(values)/4, np.mean(values)) if np.argmin(values)>len(values)/2 else
            (len(values)*3/4, np.mean(values)), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="angle3,angleA=0,angleB=-90"),
            );
```

```python
# Zadanie 4
plt.fill_between(
        x = list(range(len(values))),
        y1 = values,
        where = (values >= 0),
        color= "g",
        alpha= 0.5);
plt.fill_between(
        x = list(range(len(values))),
        y1 = values,
        where = (values < 0),
        color= "r",
        alpha= 0.5);
```
